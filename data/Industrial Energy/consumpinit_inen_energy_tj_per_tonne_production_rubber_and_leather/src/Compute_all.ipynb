```python
# Main process

# Read Input Data and Select USA Data

df_IEA = pd.read_excel('../row_data/IEA - Energy Efficiency Indicators database - Extended (December 2022).xlsb',
                        engine='pyxlsb',sheet_name='Industry - Energy',skiprows=[0])

df_IEA_USA = df_IEA[df_IEA['Country']=='United States'].loc[df_IEA['Product'].str.startswith('Total final energy use (PJ)')]

df_IPPU = pd.read_csv('../row_data/INITIAL_INDUSTRIAL_PRODUCTION.csv')

df_IPPU_USA = df_IPPU[df_IPPU['iso_alpha_3'] == 'USA']

# Cross Walk with IEA data:
industries_correspondence = {
  'consumpinit_inen_energy_tj_per_tonne_production_electronics' : [(0.5, 'Machinery [ISIC 25-28]')],
  'consumpinit_inen_energy_tj_per_tonne_production_plastic' : [(0.5, 'Rubber and plastic [ISIC 22]')],
  'consumpinit_inen_energy_tj_per_tonne_production_chemicals' : [(1.0, 'Chemicals and chemical products [ISIC 20-21]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_cement' : [(0.75, 'Non-metallic minerals [ISIC 23]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_glass' : [(0.125, 'Non-metallic minerals [ISIC 23]')], 
  'consumpinit_inen_energy_tj_per_mmm_gdp_other_product_manufacturing' : [(1, 'Construction [ISIC 41-43]'),
                                                                        (1,'Food and tobacco [ISIC 10-12]'),
                                                                         (0.5,'Machinery [ISIC 25-28]'),
                                                                         (1,'Non-specified manufacturing'),
                                                                         (1,'Other manufacturing [ISIC 31-32]'),
                                                                         (1,'Transport equipment [ISIC 29-30]')],
  'consumpinit_inen_energy_total_pj_agriculture_and_livestock' : [(1.0, 'Agriculture forestry and fishing [ISIC 01-03]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_textiles' : [(0.5, 'Textiles and leather [ISIC 13-15]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_rubber_and_leather' : [(0.5, 'Rubber and plastic [ISIC 22]'),
                                                                         (0.5, 'Textiles and leather [ISIC 13-15]')],
  'consumpinit_inen_energy_tj_per_tonne_production_metals' : [(1.0, 'Basic metals [ISIC 24]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_lime_and_carbonite' : [(0.125, 'Non-metallic minerals [ISIC 23]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_paper' : [(1.0, 'Paper pulp and printing [ISIC 17-18]')],
  'consumpinit_inen_energy_tj_per_tonne_production_wood' : [(1.0, 'Wood and wood products [ISIC 16]')]
}
# Cross Walk with IPPU data
ippu_correspondence = {
  'consumpinit_inen_energy_tj_per_tonne_production_electronics' : 'prodinit_ippu_electronics_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_plastic' : 'prodinit_ippu_plastic_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_chemicals' : 'prodinit_ippu_chemicals_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_cement' : 'prodinit_ippu_cement_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_glass' : 'prodinit_ippu_glass_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_textiles' : 'prodinit_ippu_textiles_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_rubber_and_leather' : 'prodinit_ippu_rubber_and_leather_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_metals' : 'prodinit_ippu_metals_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_lime_and_carbonite' : 'prodinit_ippu_lime_and_carbonite_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_paper' : 'prodinit_ippu_paper_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_wood' : 'prodinit_ippu_paper_tonne'
}

years_hist = range(2010,2020+1)


def keep_numbers(arr):
    return [x for x in arr if isinstance(x, (int, float))]
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

for var_name in industries_correspondence.keys():
    fraction_subsector_vect = industries_correspondence[var_name]
    if var_name in ippu_correspondence.keys():
        IPPU_var = ippu_correspondence[var_name]
        df_ippu_all=df_IPPU_USA[['iso_alpha_3','year',IPPU_var]]


    # Build fraction-subsector Dataframe with total
    
    merged_df=pd.DataFrame()
    merged_df['Year']=years_hist
    merged_df['Nation']='United States'
    
    subsector_vect=[]
    frac_vect=[]
    
    for frc_sub in fraction_subsector_vect:
        (fraction,subsector)=frc_sub
        
        subsector=subsector.replace(':','_')
        subsector=subsector.replace(';','_')

        #print(subsector)
        
        subsector_vect.append(subsector)
        frac_vect.append(fraction)
        
        df_enduse = df_IEA_USA.loc[df_IEA_USA['Subsector']==subsector]
        # df_enduse = df_enduse[df_enduse['Product'].str.startswith('Total final energy use (PJ)')]
        
       
        nation_enduse = df_enduse['Country'].values
        
        c_y_end=[]
        for c in nation_enduse:
            for y in years_hist:
                
                end_use = df_enduse.loc[(df_enduse['Country']==c)][y].values[0]

                if is_number(end_use):
                    end_use=float(end_use)

                    c_y_end.append([c,y,end_use])
        
        subsector_df=pd.DataFrame(c_y_end,columns=['Nation','Year',subsector+'_Total_(PJ)'])
        
        merged_df=pd.merge(merged_df, subsector_df, on=['Nation', 'Year'])
        
    merged_df['Total'] = 0 
    for i in range(len(subsector_vect)):
        subsector=subsector_vect[i]
        frac=frac_vect[i]
        merged_df['Total'] = merged_df['Total'] +frac * merged_df[subsector+'_Total_(PJ)']

    # if var_name=='consumpinit_inen_energy_tj_per_tonne_production_rubber_and_leather':
    #     print(merged_df)

    year_dato=[]
        
    for year in years_hist:
            
        PJ=merged_df.loc[merged_df['Year']==year]['Total'].values[0]

        if var_name in ippu_correspondence.keys():
            tonne = df_IPPU_USA.loc[df_IPPU_USA['year']==year][IPPU_var].values[0]
            
            if tonne!=0:
                dato=PJ*1e3/tonne
            else:
                dato=0

        if var_name=='consumpinit_inen_energy_total_pj_agriculture_and_livestock':
            dato=PJ

        if var_name=='consumpinit_inen_energy_tj_per_mmm_gdp_other_product_manufacturing':
            pass
            # Need GDP for Louisiana
            # gdp = df_gdp_all.loc[(df_gdp_all['Nation']==count)&(df_gdp_all['Year']==year)]['gdp_mmm_usd'].values[0]
            # dato=PJ*1e3/gdp

    

        year_dato.append(['LA','louisiana',year,dato])
        
    df_historical_out = pd.DataFrame(year_dato, columns=['iso_code3','Region', 'Year',var_name])

    if var_name!='consumpinit_inen_energy_tj_per_mmm_gdp_other_product_manufacturing' and var_name!='consumpinit_inen_energy_total_pj_agriculture_and_livestock':

        df_historical_out.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/historical/'+var_name+'.csv')

        years_proj=range(max(years_hist)+1,2051)
        
        year_dato_proj=[]
        
        
        for year in years_proj:
            
            dato=df_historical_out.loc[df_historical_out['Year']==max(years_hist)][var_name].values[0]
            
            year_dato_proj.append(['LA','louisiana',year,dato])
                
        df_projected_out = pd.DataFrame(year_dato_proj, columns=['iso_code3','Region', 'Year',var_name])
                    
        df_projected_out.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/projected/'+var_name+'.csv')
        
        print('Ready: ', var_name)


        
```
