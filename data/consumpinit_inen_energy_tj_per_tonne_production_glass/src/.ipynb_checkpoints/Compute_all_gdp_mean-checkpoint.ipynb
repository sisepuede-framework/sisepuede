```python
#Socioeconomic varibles

#GDP
df_gdp_hist = pd.read_csv(ggpdir+'/Socioeconomic/gdp_mmm_usd/input_to_sisepuede/historical/gdp_mmm_usd.csv')
df_gdp_proj = pd.read_csv(ggpdir+'/Socioeconomic/gdp_mmm_usd/input_to_sisepuede/projected/gdp_mmm_usd.csv')

nation_gdp = sorted(set(df_gdp_proj['Nation'].values))

years_all = range(min(df_gdp_hist['Year'].values),max(df_gdp_proj['Year'].values)+1)
c_y_gdp = []
for c in nation_gdp:
    for y in years_all:
    
        try:
            gdp = df_gdp_hist.loc[(df_gdp_hist['Nation']==c) & (df_gdp_hist['Year']==y),]['gdp_mmm_usd'].values[0]
            
        except:
            gdp = df_gdp_proj.loc[(df_gdp_proj['Nation']==c) & (df_gdp_proj['Year']==y)]['gdp_mmm_usd'].values[0]
            
        c_y_gdp.append([c,y,gdp])

df_gdp_all = pd.DataFrame(c_y_gdp,columns=['Nation', 'Year', 'gdp_mmm_usd'])



#Rural population
df_popru_hist = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_rural/input_to_sisepuede/historical/population_gnrl_rural.csv')
df_popru_proj = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_rural/input_to_sisepuede/projected/population_gnrl_rural.csv')

#Urban population
df_popur_hist = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_urban/input_to_sisepuede/historical/population_gnrl_urban.csv')
df_popur_proj = pd.read_csv(ggpdir+
        '/Socioeconomic/population_gnrl_urban/input_to_sisepuede/projected/population_gnrl_urban.csv')

nation_pop = sorted(set(df_popur_proj['Nation'].values))

years_all = range(min(df_popur_hist['Year'].values),max(df_popur_proj['Year'].values)+1)
c_y_pop = []
for c in nation_pop:
    for y in years_all:
    
        try:
            pop_ur = df_popur_hist.loc[(df_popur_hist['Nation']==c) &
                                       (df_popur_hist['Year']==y),]['population_gnrl_urban'].values[0]
            pop_ru = df_popru_hist.loc[(df_popru_hist['Nation']==c) &
                                       (df_popru_hist['Year']==y),]['population_gnrl_rural'].values[0]
            
        except:
            pop_ur = df_popur_proj.loc[(df_popur_proj['Nation']==c) &
                                       (df_popur_proj['Year']==y),]['population_gnrl_urban'].values[0]
            pop_ru = df_popru_proj.loc[(df_popru_proj['Nation']==c) &
                                       (df_popru_proj['Year']==y),]['population_gnrl_rural'].values[0]
            
        c_y_pop.append([c,y,pop_ur,pop_ru])

df_pop_all = pd.DataFrame(c_y_pop,columns=['Nation', 'Year', 'population_gnrl_urban','population_gnrl_rural'])

#All Together
nation_pop_gdp = sorted(set(nation_pop).intersection(set(nation_gdp)))
years_all = range(min(df_popur_hist['Year'].values),max(df_popur_proj['Year'].values)+1)
c_y_pop_gdp = []
for c in nation_pop_gdp:
    iso3 = df_gdp_hist.loc[df_gdp_hist['Nation']==c]['iso_code3'].values[0]
    for y in years_all:
        pop_ur = df_pop_all.loc[(df_pop_all['Nation']==c) &
                                       (df_pop_all['Year']==y),]['population_gnrl_urban'].values[0]
        
        pop_ru = df_pop_all.loc[(df_pop_all['Nation']==c) &
                                       (df_pop_all['Year']==y),]['population_gnrl_rural'].values[0]
        
        gdp = df_gdp_all.loc[(df_gdp_all['Nation']==c) & (df_gdp_all['Year']==y)]['gdp_mmm_usd'].values[0]
        
        c_y_pop_gdp.append([c,iso3,y,pop_ur,pop_ru,gdp])
df_pop_gdp_all = pd.DataFrame(c_y_pop_gdp,columns=['Nation','iso_code3', 'Year',
                                                   'population_gnrl_urban',
                                                   'population_gnrl_rural',
                                                   'gdp_mmm_usd'])
df_pop_gdp_all['bias']=1
df_pop_gdp_all['gdp_usd_per_cap']=df_pop_gdp_all.apply(lambda row: 
     1e9*row['gdp_mmm_usd']/(row['population_gnrl_rural']+row['population_gnrl_urban']),axis=1)

df_pop_gdp_all
```

```python
#IEA DATA
# df_IEA=pd.read_csv("../row_data/IEA_Energy_Industry.csv")
df_IEA=pd.read_excel('../row_data/IEA - Energy Efficiency Indicators database - Extended (December 2022).xlsb',
                        engine='pyxlsb',sheet_name='Industry - Energy',skiprows=[0])

#Cross Walk with IEA data:
industries_correspondence = {
  'consumpinit_inen_energy_tj_per_tonne_production_electronics' : [(0.5, 'Machinery [ISIC 25-28]')],
  'consumpinit_inen_energy_tj_per_tonne_production_plastic' : [(0.5, 'Rubber and plastic [ISIC 22]')],
  'consumpinit_inen_energy_tj_per_tonne_production_chemicals' : [(1.0, 'Chemicals and chemical products [ISIC 20-21]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_cement' : [(0.75, 'Non-metallic minerals [ISIC 23]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_glass' : [(0.125, 'Non-metallic minerals [ISIC 23]')], 
  'consumpinit_inen_energy_tj_per_mmm_gdp_other_product_manufacturing' : [(1, 'Construction [ISIC 41-43]'),
                                                                        (1,'Food and tobacco [ISIC 10-12]'),
                                                                         (0.5,'Machinery [ISIC 25-28]'),
                                                                         (1,'Non-specified manufacturing'),
                                                                         (1,'Other manufacturing [ISIC 31-32]'),
                                                                         (1,'Transport equipment [ISIC 29-30]')],
#                                                                          (1,'Wood and wood products [ISIC 16]')], 
  'consumpinit_inen_energy_total_pj_agriculture_and_livestock' : [(1.0, 'Agriculture forestry and fishing [ISIC 01-03]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_textiles' : [(0.5, 'Textiles and leather [ISIC 13-15]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_rubber_and_leather' : [(0.5, 'Rubber and plastic [ISIC 22]'),
                                                                         (0.5, 'Textiles and leather [ISIC 13-15]')],
  'consumpinit_inen_energy_tj_per_tonne_production_metals' : [(1.0, 'Basic metals [ISIC 24]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_lime_and_carbonite' : [(0.125, 'Non-metallic minerals [ISIC 23]')], 
  'consumpinit_inen_energy_tj_per_tonne_production_paper' : [(1.0, 'Paper pulp and printing [ISIC 17-18]')],
  'consumpinit_inen_energy_tj_per_tonne_production_wood' : [(1.0, 'Wood and wood products [ISIC 16]')]
}
#IPPU data
ippu_correspondence = {
  'consumpinit_inen_energy_tj_per_tonne_production_electronics' : 'prodinit_ippu_electronics_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_plastic' : 'prodinit_ippu_plastic_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_chemicals' : 'prodinit_ippu_chemicals_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_cement' : 'prodinit_ippu_cement_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_glass' : 'prodinit_ippu_glass_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_textiles' : 'prodinit_ippu_textiles_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_rubber_and_leather' : 'prodinit_ippu_rubber_and_leather_tonne', 
  'consumpinit_inen_energy_tj_per_tonne_production_metals' : 'prodinit_ippu_metals_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_lime_and_carbonite' : 'prodinit_ippu_lime_and_carbonite_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_paper' : 'prodinit_ippu_paper_tonne',
  'consumpinit_inen_energy_tj_per_tonne_production_wood' : 'prodinit_ippu_paper_tonne'
}

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

years_hist = range(min(df_pop_gdp_all['Year'].values),2020+1)

for var_name in industries_correspondence.keys():
    fraction_subsector_vect = industries_correspondence[var_name]
    if var_name in ippu_correspondence.keys():
        IPPU_var = ippu_correspondence[var_name]
        df_ippu_hist = pd.read_csv(ggpdir+'/IPPU/'+IPPU_var+'/input_to_sisepuede/historical/'+IPPU_var+'.csv')
        df_ippu_proj = pd.read_csv(ggpdir+'/IPPU/'+IPPU_var+'/input_to_sisepuede/projected/'+IPPU_var+'.csv')
        df_ippu_all=pd.concat([df_ippu_hist,df_ippu_proj])
        df_ippu_all = df_ippu_all.replace([np.nan, np.inf, np.finfo(np.float64).max], 0)
        
    years_all = range(min(df_pop_gdp_all['Year'].values),2050+1)

    ISO3 = sorted(set(df_pop_gdp_all['iso_code3'].values).intersection(set(df_ippu_all['location_code'].values)))

    coun_year=[]

    for iso3 in ISO3:
        count = df_gdp_hist.loc[df_gdp_hist['iso_code3']==iso3]['Nation'].values[0]
        for year in years_hist:
            
            coun_year.append([count,iso3,year])

    merged_df=pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
    
    subsector_vect=[]
    frac_vect=[]
    
    for frc_sub in fraction_subsector_vect:
        (fraction,subsector)=frc_sub
        subsector=subsector.replace(':','_')
        subsector=subsector.replace(';','_')
        
        subsector_vect.append(subsector)
        frac_vect.append(fraction)
        
        df_enduse = df_IEA.loc[df_IEA['Subsector']==subsector]
        df_enduse = df_enduse[df_enduse['Product'].str.startswith('Total final energy use (PJ)')]
        
       
        nation_enduse = df_enduse['Country'].values
        
        c_y_end=[]
        for c in nation_enduse:
            for y in years_hist:
                
                end_use = df_enduse.loc[(df_enduse['Country']==c)][y].values[0]

                if is_number(end_use):
                    end_use=float(end_use)

                    c_y_end.append([c,y,end_use])
        
        subsector_df=pd.DataFrame(c_y_end,columns=['Nation','Year',subsector+'_Total_(PJ)'])
        
        merged_df=pd.merge(merged_df, subsector_df, on=['Nation', 'Year'])
        
    merged_df['Total'] = 0 
    for i in range(len(subsector_vect)):
        subsector=subsector_vect[i]
        frac=frac_vect[i]
        merged_df['Total'] = merged_df['Total'] +frac * merged_df[subsector+'_Total_(PJ)']
    

    ISO3 = sorted(set(df_pop_gdp_all['iso_code3'].values).intersection(set(df_ippu_all['location_code'].values)))


    coun_year_dato_1=[]

    for iso3 in ISO3:
        count = df_gdp_hist.loc[df_gdp_hist['iso_code3']==iso3]['Nation'].values[0]
        for year in years_hist:
            
            if (count,year) in zip(merged_df['Nation'].values,merged_df['Year'].values):
            
                PJ=merged_df.loc[(merged_df['iso_code3']==iso3)&
                                (merged_df['Year']==year)]['Total'].values[0]

                if var_name in ippu_correspondence.keys():
                    tonne = df_ippu_all.loc[(df_ippu_all['location_code']==iso3)&
                                    (df_ippu_all['year']==year)][IPPU_var].values[0]
                    if tonne!=0:
                        dato=PJ*1e3/tonne
                    else:
                        dato=0

                if var_name=='consumpinit_inen_energy_total_pj_agriculture_and_livestock':
                    dato=PJ

                if var_name=='consumpinit_inen_energy_tj_per_mmm_gdp_other_product_manufacturing':
                    gdp = df_gdp_all.loc[(df_gdp_all['Nation']==count)&(df_gdp_all['Year']==year)]['gdp_mmm_usd'].values[0]
                    dato=PJ*1e3/gdp
           
                
                coun_year_dato_1.append([count,iso3,year,dato])

    df_historical_1 = pd.DataFrame(coun_year_dato_1, columns=['Nation','iso_code3', 'Year',var_name])
    
    coun_year_dato=[]
    
    for iso3 in ISO3:
        count = df_gdp_hist.loc[df_gdp_hist['iso_code3']==iso3]['Nation'].values[0]
        for year in years_hist:
            
            if (count,year) in zip(merged_df['Nation'].values,merged_df['Year'].values):
            
                PJ=merged_df.loc[(merged_df['iso_code3']==iso3)&
                                (merged_df['Year']==year)]['Total'].values[0]

                if var_name in ippu_correspondence.keys():
                    tonne = df_ippu_all.loc[(df_ippu_all['location_code']==iso3)&
                                    (df_ippu_all['year']==year)][IPPU_var].values[0]
                    if tonne!=0:
                        dato=PJ*1e3/tonne
                    else:
                        dato=0

                if var_name=='consumpinit_inen_energy_total_pj_agriculture_and_livestock':
                    dato=PJ

                if var_name=='consumpinit_inen_energy_tj_per_mmm_gdp_other_product_manufacturing':
                    gdp = df_gdp_all.loc[(df_gdp_all['Nation']==count)&(df_gdp_all['Year']==year)]['gdp_mmm_usd'].values[0]
                    dato=PJ*1e3/gdp
            else:
                
                gdp_per_cap=df_pop_gdp_all.loc[(df_pop_gdp_all['Nation']==count)&
                        (df_pop_gdp_all['Year']==year)]['gdp_usd_per_cap'].values[0]

                var_per_gdp_per_cap=df_energy_tonne_mean.loc[df_energy_tonne_mean['Year']==y][
                    var_name+'_mean_per_gdp_percap'].values[0]
                
                
                
                dato=var_per_gdp_per_cap*gdp_per_cap
                
                max_year=max(df_historical_1[df_historical_1['Year']==y][var_name].values)
                
                min_year=min(df_historical_1[df_historical_1['Year']==y][var_name].values)
                
#                 print(var_name,max_year)
                
                if dato>max_year:
                    dato=max_year
                if dato<min_year:
                    dato=min_year
                
            coun_year_dato.append([count,iso3,year,dato])
            
    df_historical_out = pd.DataFrame(coun_year_dato, columns=['Nation','iso_code3', 'Year',var_name])
    
#     df_historical_out
    
#     if var_name=='consumpinit_inen_energy_tj_per_tonne_production_metals':
#         stop

    df_historical_out.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/historical/'+var_name+'.csv')
    
    
    years_proj=range(max(years_hist)+1,2051)
    
    coun_year_dato_proj=[]
    
    for iso3 in ISO3:
        count = df_gdp_hist.loc[df_gdp_hist['iso_code3']==iso3]['Nation'].values[0]
        for year in years_proj:
            
            dato=df_historical_out.loc[(df_historical_out['Nation']==count)&
                                       (df_historical_out['Year']==max(years_hist))
                                      ][var_name].values[0]
            
            coun_year_dato_proj.append([count,iso3,year,dato])
            
    df_projected_out = pd.DataFrame(coun_year_dato_proj, columns=['Nation','iso_code3', 'Year',var_name])
                
    df_projected_out.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/projected/'+var_name+'.csv')
    
    print('Ready: ', var_name)
```
