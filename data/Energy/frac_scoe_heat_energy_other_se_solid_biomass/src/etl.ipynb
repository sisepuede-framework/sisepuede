```python
import pandas as pd
import numpy as np
```

```python
# Loading the baseline
df = pd.read_excel('../raw_data/World Energy Balances Highlights 2023.xlsx',sheet_name=3)
df.head()
```

```python
# Fix the column names
df.columns = df.iloc[0]
df = df.iloc[1:]
df.head()
```

```python
# some data cleaning
df.replace(to_replace='..', value=0,inplace=True)
df.head()
```

```python
#Select jus US data
df = df[df['Country']=='United States']
df.head()
```

```python
# select just other SE
df=df[df['Flow']=='Other final consumption (PJ)']
df.head()
```

```python
# Now we can load the proportions
df_prop=pd.read_excel('../raw_data/frac_scoe_heat_energy_proportion.xlsx')
df_prop.head()
```

```python
np.unique(df_prop['SISEPUEDE_var'].values)
```

```python
df_prop_var = df_prop[df_prop['SISEPUEDE_var']=='frac_scoe_heat_energy_other_se_solid_biomass']
fuel       = df_prop_var['WEBH2021_var'].values[0]
p = df_prop_var['Proportion'].values[0]
```

```python
years=list(range(2015,2022))
data = []
for year in years:
    f = df[df['Product']==fuel][year].values[0]
    t = df[df['Product']=='Total'][year].values[0]
    data.append(p*f/t)
data
```

```python
pd.DataFrame(zip(years,data)).to_csv('../new_input.csv',index=False,header=False)
```
