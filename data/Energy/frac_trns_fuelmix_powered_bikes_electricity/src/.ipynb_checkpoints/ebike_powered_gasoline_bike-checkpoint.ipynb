## Obtencin de las fracciones de energa utilizada

De acuerdo al siguiente [recurso](https://ieeexplore.ieee.org/abstract/document/9179285?casa_token=7c6CFavQo_MAAAAA:MfI4jd9z6gvMZ7T5R5MyJbB1bmdanV9fEqJmCFIwBhvFIc0YbfPzmXao2jao4F0CZqkf3fwQNCo), 



Suppose motorbike speed should be 50Km/h with an acceleration of $1.16ms^2$. Approximately 2093 watt hub motor should be placed within a bike tire.

At rated angular speed power consumed by machine is, 2080 watt.

### Ventas unitarias de motorbikes en el mundo



```python
## Ventas unitarias de motorbikes en el mundo
## https://www.statista.com/outlook/mmo/motorcycles/worldwide#unit-sales

motorbikes_all = {
    "on_road_motorcyles" : {"year" : list(range(2014,2028)), 
                            "value" : [28.4, 27.79, 28.41, 30.21, 30.55, 31.07, 28.69, 31.41, 29.65,
                                      30.71, 34.15, 35.58, 37.05, 38.65]}
}
```

### Ventas unitarias de ebikes en el mundo (proyectadas)


```python
### Ventas unitarias de ebikes en el mundo (proyectadas)
## https://www.statista.com/statistics/1334665/global-e-bike-market-volume-forecast/#:~:text=In%202022%2C%20the%20global%20e,reach%20around%2036.5%20million%20units.

ebikes_projected = {"year" : list(range(2022, 2031)),
                    "value" : [36_463.58, 40_051.59, 43_992.67, 48_321.55, 53_076.39, 58_299.11,
                              64_035.74, 70_336.65, 77_258]}
```

### Ventas unitarias de ebikes en el mundo (proyectadas)


```python
### Ventas unitarias de ebikes en el mundo (proyectadas)
## https://www.tandfonline.com/doi/full/10.1080/01441647.2015.1069907?casa_token=V5xX4lNwLC0AAAAA%3A4GNLL5mJ9Ys2kjxTSRWgpHdGpy12niHt_oYHeRv__kcdeiF6fAjJgsGjFNba0Vapg68fh35lLzcVehM
import pandas as pd

ebikes_historical = pd.read_csv("../raw_data/T0002-10.1080_01441647.2015.1069907.csv")
ebikes_historical.rename(columns = {ebikes_historical.columns[0] : "Nation"}, inplace = True)
ebikes_historical
```

```python
## Convertimos datos a data frames
melt_ebikes_historical = ebikes_historical.melt(id_vars="Nation", value_vars = ebikes_historical.columns[1:])
total_ebikes_historical = melt_ebikes_historical.query("Nation == 'Total'").reset_index(drop=True)[["variable", "value"]]
total_ebikes_historical.rename(columns = {"variable":"year"}, inplace = True)
total_ebikes_historical["value"] = total_ebikes_historical["value"].apply(lambda x : float(x.encode('ascii', 'ignore'))/1_000_000)
total_ebikes_historical
```

```python
total_ebikes_projected = pd.DataFrame(ebikes_projected )
total_ebikes_projected["value"] = total_ebikes_projected["value"]/1000
total_ebikes_projected
```

```python
## Concatenamos valores histricos y proyectados
import numpy as np

missing_data = pd.DataFrame([(i,np.nan) for i in range(2016, 2022)], columns = ["year", "value"])
total_ebikes = pd.concat([total_ebikes_historical, missing_data, total_ebikes_projected])
total_ebikes["value"] = total_ebikes["value"].interpolate()
total_ebikes.reset_index(drop = True, inplace = True)
total_ebikes
```

```python
total_motorbikes = pd.DataFrame(motorbikes_all["on_road_motorcyles"])
total_motorbikes
```

```python
total_bikes = pd.concat([total_ebikes.rename(columns = {"value":"total_ebikes"}).set_index("year"), 
           total_motorbikes.rename(columns = {"value" : "total_motorbikes"}).set_index("year")], axis = 1).reset_index()
total_bikes.iloc[2:14].plot("year", ["total_ebikes", "total_motorbikes"])

import matplotlib.pyplot as plt

plt.ylabel("Ventas\n(millones)")
plt.legend()
plt.savefig("ventas_ebikes_motorbikes.png")
```

### Obtenemos el consumo de energa en watts
#### E-bikes

```python
# Approximately 2093 watt hub motor should be placed within a bike tire.
# At rated angular speed power consumed by machine is, 2080 watt.

scalar_ebikes = 2080
scalar_motorbikes = 2093
total_ebikes["year"] = total_ebikes["year"].apply(lambda x: int(x))
total_ebikes_watts = total_ebikes.query("year <= 2022")
total_ebikes_watts["value"] = total_ebikes_watts["value"]*scalar_ebikes

total_ebikes_watts.rename(columns = {"value" : "ebikes_watts"}, inplace = True)
total_ebikes_watts
```

#### Motorbikes


```python
total_motorbikes_watts = pd.concat([pd.DataFrame([(i, np.nan) for i in range(2011, 2014)], columns = ["year", "value"]),
            total_motorbikes]).query("year <=2022").interpolate().fillna(method='bfill')

total_motorbikes_watts["value"] = total_motorbikes_watts["value"] * scalar_motorbikes

total_motorbikes_watts = total_motorbikes_watts.rename(columns = {"value" : "motorbikes_watts"}).reset_index(drop = True)
total_motorbikes_watts
```

#### Merge

```python
total_watts = pd.concat([total_ebikes_watts.set_index("year"), 
                         total_motorbikes_watts.set_index("year")], axis = 1).reset_index()
total_watts["total"] = total_watts["ebikes_watts"] + total_watts["motorbikes_watts"]
total_watts
```

```python
frac_watts = total_watts[["year"]]
frac_watts["frac_trns_fuelmix_powered_bikes_electricity"] = total_watts["ebikes_watts"] /total_watts["total"]
frac_watts["frac_trns_fuelmix_powered_bikes_gasoline"] = total_watts["motorbikes_watts"] /total_watts["total"]
frac_watts
```

```python
frac_watts.to_csv("../raw_data/frac_trns_fuelmix_powered_bikes.csv", index = False)
```
