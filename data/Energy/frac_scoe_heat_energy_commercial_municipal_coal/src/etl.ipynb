```python
df = pd.read_csv('../raw_data/heat_energy.csv')  
df.head()
```

```python
df['total'] = df.iloc[:,1:].sum(axis=1)
df.head()
```

```python
df[['coal','natural_gas','diesel','hydrocarbon_gas_liquids','kerosene','gasoline','solid_biomass','electricity']] = df[['coal','natural_gas','diesel','hydrocarbon_gas_liquids','kerosene','gasoline','solid_biomass','electricity']].div(df['total'], axis=0)
df.head()
```

```python
df[["year","coal"]].to_csv('../new_input.csv', index=False,header=False)
```
