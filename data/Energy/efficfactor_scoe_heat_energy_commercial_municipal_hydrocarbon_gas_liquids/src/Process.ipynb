```python
import pandas as pd
import numpy as np
import os
import sys
import inspect

parentdir = os.path.abspath('..')
parentdir_name = os.path.basename(parentdir)

grandparentdir=os.path.dirname(parentdir)
ggpdir=os.path.dirname(grandparentdir)

sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)
sys.path.insert(0, ggpdir)
#Variable name
var_name=parentdir_name
var_name
```

```python
# As stated in https://doi.org/10.1016/j.enconman.2024.118284, Table 4 (last visited 10/July/2024) 
# the gas boiler efficiency for Central Heating is 0.94. This will be the historical and projected value.


#### HISTORICAL ######
dato=0.94
years_hist=np.arange(2010,2024,1)
coun_year=[]
for year in years_hist:
    coun_year.append(['LA','louisiana',year,dato])
df_historical_out = pd.DataFrame(coun_year,columns=['iso_code3','Region', 'Year',var_name])
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
df_historical_out
```

```python
#Variable name
#### projected ######
years_porj=np.arange(max(years_hist)+1,2051,1)
coun_year=[]

for year in years_porj:
    coun_year.append(['LA','louisiana',year,dato])
df_projected_out = pd.DataFrame(coun_year, columns=['iso_code3','Region', 'Year',var_name])
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
df_projected_out
```
