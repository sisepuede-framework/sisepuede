```python
import pandas as pd
import numpy as np
```

```python
# Visualize input data
# Downloaded from https://www.bts.gov/browse-statistical-products-and-data/state-transportation-statistics/freight-flows-state (last visited 1/June/2024)

df_freight_all = pd.read_csv("../row_data/state total.csv")
df_freight_all.head()
```

```python
df_freight_louisiana = df_freight_all[df_freight_all['Unnamed: 0']== 'Louisiana']
df_freight_louisiana
```

```python
# Clean data

years_hist = range(2010,2022+1)
year_data = []
for year  in years_hist:
    
    if year <= 2018:
        val = df_freight_louisiana['Ton Miles (millions)'].values[0]
    if year == 2019:
        val = df_freight_louisiana['Ton Miles (millions).1'].values[0]
    if year == 2020:
        val = df_freight_louisiana['Ton Miles (millions).2'].values[0]
    if year == 2021:
        val = df_freight_louisiana['Ton Miles (millions).3'].values[0]
    if year == 2022:
        val = df_freight_louisiana['Ton Miles (millions).4'].values[0]

    year_data.append([year,val])

df_freight_louisiana_clean = pd.DataFrame(year_data, columns = ['Year','Tonne_Miles (millions)'])
df_freight_louisiana_clean


# Convert Tonne/Miles to Tonne/Km: tonne-kilometres =  1/1.6352 ton-mile.


df_freight_louisiana_clean['deminit_trde_freight_mt_km'] = 1/1.6352*(df_freight_louisiana_clean['Tonne_Miles (millions)'])

df_freight_louisiana_clean
```

```python
df_historical = df_freight_louisiana_clean[['Year','deminit_trde_freight_mt_km']]
df_historical['iso_code3'] = 'LA'
df_historical['Region'] = 'louisiana'
df_historical
```

```python
#Projected data will be taken as the last historical value

years_proj = range(max(years_hist)+1,2050+1)
df_projected = pd.DataFrame(years_proj,columns=['Year'])
df_projected['iso_code3'] = 'LA'
df_projected['Region'] = 'louisiana'
df_projected['deminit_trde_freight_mt_km'] = 141.863381
df_projected
```

```python
# Write data
var = 'deminit_trde_freight_mt_km'
df_historical.to_csv(grandparentdir+'/'+var+'/input_to_sisepuede/historical/'+var+'.csv')
df_projected.to_csv(grandparentdir+'/'+var+'/input_to_sisepuede/projected/'+var+'.csv')
```
