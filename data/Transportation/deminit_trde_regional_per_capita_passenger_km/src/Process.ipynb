```python
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)

var_name=os.path.basename(parentdir)

var_name
```

```python
import pandas as pd
import numpy as np
```

```python
# Regional data is the national value
# Obtained from https://data.oecd.org/transport/passenger-transport.htm#indicator-chart (last visited 01/June/2024)

df_pass=pd.read_csv("../row_data/DP_LIVE_18012023221837518.csv")
df_pass=df_pass.fillna(0)
df_pass
```

```python
df_pass_USA = df_pass[df_pass['LOCATION']=='USA']
df_pass_USA
```

```python
# historical data

years_hist = range(2010,2017+1)
df_pass_USA_hist = pd.DataFrame(years_hist,columns=['Year'])

df_pass_USA_hist ['RAIL'] = df_pass_USA[(df_pass_USA['TIME']>= 2010) & 
                        (df_pass_USA['TIME']<= 2017) &
                        (df_pass_USA['SUBJECT']== 'RAIL')]['Value'].values

df_pass_USA_hist ['ROAD'] = df_pass_USA[(df_pass_USA['TIME']>= 2010) & 
                        (df_pass_USA['TIME']<= 2017) &
                        (df_pass_USA['SUBJECT']== 'ROAD')]['Value'].values

df_pass_USA_hist ['INLAND'] = df_pass_USA[(df_pass_USA['TIME']>= 2010) & 
                        (df_pass_USA['TIME']<= 2017) &
                        (df_pass_USA['SUBJECT']== 'INLAND')]['Value'].values

df_pass_USA_hist ['Total_PKm'] = df_pass_USA_hist['RAIL'] + df_pass_USA_hist['ROAD'] + df_pass_USA_hist['INLAND'] 

df_pass_USA_hist
```

```python
# Population data

df_hist_pop_rural = pd.read_csv('../../../SocioEconomic/population_gnrl_rural/input_to_sisepuede/historical/population_gnrl_rural.csv')
df_proj_pop_rural = pd.read_csv('../../../SocioEconomic/population_gnrl_rural/input_to_sisepuede/projected/population_gnrl_rural.csv')
df_pop_rural_all = pd.concat([df_hist_pop_rural,df_proj_pop_rural])

df_hist_pop_urban = pd.read_csv('../../../SocioEconomic/population_gnrl_urban/input_to_sisepuede/historical/population_gnrl_urban.csv')
df_proj_pop_urban = pd.read_csv('../../../SocioEconomic/population_gnrl_urban/input_to_sisepuede/projected/population_gnrl_urban.csv')
df_pop_urban_all = pd.concat([df_hist_pop_urban,df_proj_pop_urban])

df_pop_all = df_pop_rural_all.copy()
df_pop_all['population_gnrl_urban'] = df_pop_urban_all['population_gnrl_urban']
df_pop_all['population_all'] = df_pop_all['population_gnrl_urban'] + df_pop_all['population_gnrl_rural']

df_pop_all
```

```python
# Merge data

df_all_hist = df_pop_all[(df_pop_all['Year']>= 2010) & (df_pop_all['Year']<= 2017)]
df_all_hist['Total_PKm'] = df_pass_USA_hist['Total_PKm'].values

df_all_hist[var_name] = df_all_hist['Total_PKm']*1e6/df_all_hist['population_all']

df_all_hist
```

```python
df_historical = df_all_hist[['Region','iso_code3','Year',var_name]]
df_historical
```

```python
#Projected data will be taken as the last historical value

years_proj = range(max(years_hist)+1,2050+1)
df_projected = pd.DataFrame(years_proj,columns=['Year'])
df_projected['iso_code3'] = 'LA'
df_projected['Region'] = 'louisiana'
df_projected[var_name] = df_historical[df_historical['Year']==max(years_hist)][var_name].values[0]
df_projected
```

```python
# Write data
df_historical.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/historical/'+var_name+'.csv')
df_projected.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/projected/'+var_name+'.csv')
```
