```python
# Add Parent Folder to System Path

import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)
#### In the absence of a better data we will use European indicators
# found in https://www.eea.europa.eu/data-and-maps/indicators/occupancy-rates-of-passenger-vehicles/occupancy-rates-of-passenger-vehicles (last visited 01/June/2024)

import pandas as pd
import numpy as np



#Variable name
var_name='occrate_trns_passenger_per_vehicle_aviation'
#DATAFRAMES
df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_data=pd.read_excel ('../row_data/2009 29 - Occupancy rates of passenger.xls',sheet_name='TERM29_BD_Air')
df_data
```

```python
df_data_copy=df_data
new_header = df_data.iloc[2] #grab the first row for the header
df_data_copy = df_data[3:] #take the data less the header row
df_data_copy.columns = new_header #set the header row as the df header
df_data_copy
```

```python
df_data_copy = df_data_copy[df_data_copy[['YEAR', 'PLF (decimal)']].notnull().all(1)]
df_data_copy
```

```python
#### HISTORICAL ######

years_hist = range(1991,2007+1)

df_historical = pd.DataFrame(years_hist,columns=['Year'])

#Assuming that most of the airplanes are Small narrowbody aircraft then the total of seats is 150
#source: https://web.mit.edu/airlinedata/www/Aircraft&Related.html

variable_vect=[plf*150 for plf in df_data_copy['PLF (decimal)'].values if type(plf)==float]

df_historical['occrate_trns_passenger_per_vehicle_aviation'] = variable_vect

df_historical['iso_code3'] = 'LA'
df_historical['Region'] = 'louisiana'

df_historical
```

```python
#Projected data will be taken as the last historical value
var_name='occrate_trns_passenger_per_vehicle_aviation'
years_proj = range(max(years_hist)+1,2050+1)
df_projected = pd.DataFrame(years_proj,columns=['Year'])
df_projected['iso_code3'] = 'LA'
df_projected['Region'] = 'louisiana'
df_projected[var_name] = df_historical[df_historical['Year']==max(years_hist)][var_name].values[0]
df_projected
```

```python
# Write data
df_historical.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/historical/'+var_name+'.csv')
df_projected.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/projected/'+var_name+'.csv')
```
