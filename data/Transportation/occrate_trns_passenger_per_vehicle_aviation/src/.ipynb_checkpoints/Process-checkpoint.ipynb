```python
# Add Parent Folder to System Path

import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir) 

####

import pandas as pd
import numpy as np

#Variable name
var_name='occrate_trns_passenger_per_vehicle_aviation'
#DATAFRAMES
df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_data=pd.read_excel ('../row_data/2009 29 - Occupancy rates of passenger.xls',sheet_name='TERM29_BD_Air')
df_data
```

```python
df_data_copy=df_data
new_header = df_data.iloc[2] #grab the first row for the header
df_data_copy = df_data[3:] #take the data less the header row
df_data_copy.columns = new_header #set the header row as the df header
df_data_copy
```

```python
df_data_copy = df_data_copy[df_data_copy[['YEAR', 'PLF (decimal)']].notnull().all(1)]
df_data_copy
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)

#Assuming that most of the airplanes are Small narrowbody aircraft then the total of seats is 150
#source: https://web.mit.edu/airlinedata/www/Aircraft&Related.html

years=np.unique([year for year in df_data_copy['YEAR'].values if type(year)==int])
variable_vect=[plf*150 for plf in df_data_copy['PLF (decimal)'].values if type(plf)==float]


coun_year=[]
for count in zip(countries,ISO3):
    for year,var in zip(years,variable_vect):
        coun_year.append([count[0],count[1],year,var])
        
df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year',var_name])
df_historical_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
#### PROJECTED ######

#Prediction for the next years will be the last historical value
years2=np.array([])
variable_vect2=np.array([])
for y in np.arange(max(years),2051,1):
    if y not in years:
        years2=np.append(years2,int(y))
        variable_vect2=np.append(variable_vect2,variable_vect[-1])

coun_year2=[]
for count in zip(countries,ISO3):
    for y_p in zip(years2,variable_vect2):
        coun_year2.append([count[0],count[1],y_p[0],y_p[1]])
coun_year2

df_projected_out = pd.DataFrame(coun_year2, columns=['Nation','iso_code3', 'Year',var_name])
df_projected_out
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
