```python
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)

# var_name=os.path.basename(parentdir)

# var_name
```

```python
import pandas as pd
import numpy as np
import os

# set the directory path and the prefix string
directory = grandparentdir
prefix = 'frac_trns_mtkm_dem_freight'

frac_mtkm=[]
# iterate over all items in the directory
for item in os.listdir(directory):
    # check if the item is a directory and starts with the prefix
    if os.path.isdir(os.path.join(directory, item)) and item.startswith(prefix):
        frac_mtkm.append(item)
        # print the directory name
        print(item)
frac_mtkm
```

```python
## Regional data is the national value
# Obtained from https://data.oecd.org/transport/freight-transport.htm#indicator-chart (last visited 01/June/2024)
df_freight=pd.read_csv("../row_data/DP_LIVE_18012023235438219.csv")
# df_freight=pd.read_csv("../row_data/imputated_freight_data.csv")
df_aviation=pd.read_csv("../row_data/Aviation_freigth_inputed.csv",index_col=0)

df_freight=df_freight.fillna(0)
df_freight
```

```python
df_aviation_USA = df_aviation[df_aviation['iso_code3']=='USA']

df_aviation_USA
```

```python
df_freight_USA = df_freight[df_freight['LOCATION']=='USA']
df_freight_USA
```

```python
# historical data

years_hist = range(2011,2020+1)
df_freight_USA_hist = pd.DataFrame(years_hist,columns=['Year'])

df_freight_USA_hist ['RAIL'] = df_freight_USA[(df_freight_USA['TIME']>= 2010) & 
                        (df_freight_USA['TIME']<= 2020) &
                        (df_freight_USA['SUBJECT']== 'RAIL')]['Value'].values

df_freight_USA_hist ['ROAD'] = df_freight_USA[(df_freight_USA['TIME']>= 2010) & 
                        (df_freight_USA['TIME']<= 2020) &
                        (df_freight_USA['SUBJECT']== 'ROAD')]['Value'].values

df_freight_USA_hist ['INLAND'] = df_freight_USA[(df_freight_USA['TIME']>= 2010) & 
                        (df_freight_USA['TIME']<= 2020) &
                        (df_freight_USA['SUBJECT']== 'INLAND')]['Value'].values

df_freight_USA_hist ['COAST'] = df_freight_USA[(df_freight_USA['TIME']>= 2010) & 
                        (df_freight_USA['TIME']<= 2020) &
                        (df_freight_USA['SUBJECT']== 'COAST')]['Value'].values

df_freight_USA_hist ['Aviation'] = df_aviation_USA['Aviation (mtkm)'].values

df_freight_USA_hist ['Total_MTKm'] = df_freight_USA_hist ['COAST'] + df_freight_USA_hist['RAIL'] + df_freight_USA_hist['ROAD'] + df_freight_USA_hist['INLAND'] + df_freight_USA_hist ['Aviation']

df_freight_USA_hist ['frac_trns_mtkm_dem_freight_water_borne'] = df_freight_USA_hist ['COAST']/df_freight_USA_hist ['Total_MTKm']
df_freight_USA_hist ['frac_trns_mtkm_dem_freight_road_heavy_freight'] = (df_freight_USA_hist ['ROAD']+df_freight_USA_hist ['INLAND'])/df_freight_USA_hist ['Total_MTKm']
df_freight_USA_hist ['frac_trns_mtkm_dem_freight_rail_freight'] = df_freight_USA_hist ['RAIL']/df_freight_USA_hist ['Total_MTKm']
df_freight_USA_hist ['frac_trns_mtkm_dem_freight_aviation'] = df_freight_USA_hist ['Aviation']/df_freight_USA_hist ['Total_MTKm']

df_freight_USA_hist['iso_code3'] = 'LA'
df_freight_USA_hist['Region'] = 'louisiana'

df_freight_USA_hist
```

```python
years_proj = range(max(years_hist)+1,2050+1)

for var_name in frac_mtkm:
    df_historical = df_freight_USA_hist[['Region','iso_code3','Year',var_name]]
    years_proj = range(max(years_hist)+1,2050+1)
    df_projected = pd.DataFrame(years_proj,columns=['Year'])
    df_projected['iso_code3'] = 'LA'
    df_projected['Region'] = 'louisiana'
    df_projected[var_name] = df_historical[df_historical['Year']==max(years_hist)][var_name].values[0]
    # Write data
    df_historical.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/historical/'+var_name+'.csv')
    df_projected.to_csv(grandparentdir+'/'+var_name+'/input_to_sisepuede/projected/'+var_name+'.csv')
```
