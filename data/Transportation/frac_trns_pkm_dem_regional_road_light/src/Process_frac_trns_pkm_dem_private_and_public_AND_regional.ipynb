```python
import pandas as pd
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect
import numpy as np

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
ggpdir=os.path.dirname(grandparentdir)

sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)
sys.path.insert(0, ggpdir)
```

```python
# Visualize input data
df_Passenger_Miles_USA = pd.read_excel('../row_data/table_01_40_112323.xlsx',sheet_name='1-40',skiprows=[0])

df_Passenger_Miles_USA
```

```python
mypath=grandparentdir
frac_trns_pkm_dem_private_and_public_vars = sorted([f for f in os.listdir(mypath) if f.startswith('frac_trns_pkm_dem_private_and_public')])
frac_trns_pkm_dem_private_and_public_vars
```

```python
mypath=grandparentdir
frac_trns_pkm_dem_regional_vars = sorted([f for f in os.listdir(mypath) if f.startswith('frac_trns_pkm_dem_regional')])
frac_trns_pkm_dem_regional_vars
```

```python
df_Passenger_Miles_USA['Transport']=df_Passenger_Miles_USA['Unnamed: 0']

years_hist = range(2011,2021+1)

year_data = []

for year in years_hist:

    # Private and Public data

    # Road light
    
    road_light_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Light duty vehicle, short wheel basea,b'][year].values[0]

    # Powered bikes
    
    powered_bikes_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Motorcycleb'][year].values[0]

    # Public

    public_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Busc,e'][year].values[0]

    # For walking and cycling 2017 data will be used for historical data

    walking_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Walkingj'][2017].values[0]
    
    cycling_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Cyclingj'][2017].values[0]

    human_powered_data = walking_data+cycling_data


    # Regional data

    #Aviation

    aviation_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='U.S. air carrier, certificated, domestic, all services'][year].values[0]

    # Rail includes light and heavy
    
    light_rail_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Light railf'][year].values[0]
    
    heavy_rail_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Heavy rail'][year].values[0]

    rail_data = light_rail_data + heavy_rail_data

    # Water borne

    ferry_boat_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Ferry boatg'][year].values[0]

    # Road Heavy

    motor_bus_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Motor busc,e'][year].values[0]

    commuter_bus_data = df_Passenger_Miles_USA[df_Passenger_Miles_USA['Unnamed: 0']=='Commuter buse'][year].values[0]

    road_heavy_data = public_data + motor_bus_data + commuter_bus_data

    year_data.append([year,
                      road_light_data,
                     powered_bikes_data,
                      public_data,
                      human_powered_data,
                      aviation_data,
                      rail_data,
                      road_heavy_data,
                      ferry_boat_data
                     ])

    

df_Passenger_Miles_USA_clean = pd.DataFrame(year_data,columns = ['Year', 
                                                                 'road_light_data',
                                                                 'powered_bikes_data',
                                                                 'public_data',
                                                                 'human_powered_data',
                                                                 'aviation_data',
                                                                 'rail_data',
                                                                 'road_heavy_data',
                                                                 'ferry_boat_data'])

df_Passenger_Miles_USA_clean['Total_private_and_public'] = df_Passenger_Miles_USA_clean[[
                                                                 'road_light_data',
                                                                 'powered_bikes_data',
                                                                 'public_data',
                                                                 'human_powered_data',
                                                                 'ferry_boat_data']].sum(axis=1)

df_Passenger_Miles_USA_clean['Total_regional'] = df_Passenger_Miles_USA_clean[[
                                                                 'road_light_data',
                                                                 'aviation_data',
                                                                 'rail_data',
                                                                 'road_heavy_data',
                                                                 'ferry_boat_data']].sum(axis=1)
df_Passenger_Miles_USA_clean
```

```python
df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_private_and_public_human_powered'] = df_Passenger_Miles_USA_clean['human_powered_data']/df_Passenger_Miles_USA_clean['Total_private_and_public']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_private_and_public_powered_bikes'] = df_Passenger_Miles_USA_clean['powered_bikes_data']/df_Passenger_Miles_USA_clean['Total_private_and_public']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_private_and_public_public'] = df_Passenger_Miles_USA_clean['public_data']/df_Passenger_Miles_USA_clean['Total_private_and_public']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_private_and_public_road_light'] = df_Passenger_Miles_USA_clean['road_light_data']/df_Passenger_Miles_USA_clean['Total_private_and_public']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_private_and_public_water_borne'] = df_Passenger_Miles_USA_clean['ferry_boat_data']/df_Passenger_Miles_USA_clean['Total_private_and_public']


df_Passenger_Miles_USA_clean
```

```python
df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_regional_aviation'] = df_Passenger_Miles_USA_clean['aviation_data']/df_Passenger_Miles_USA_clean['Total_regional']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_regional_rail_passenger'] = df_Passenger_Miles_USA_clean['rail_data']/df_Passenger_Miles_USA_clean['Total_regional']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_regional_road_heavy_regional'] = df_Passenger_Miles_USA_clean['road_heavy_data']/df_Passenger_Miles_USA_clean['Total_regional']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_regional_road_light'] = df_Passenger_Miles_USA_clean['road_light_data']/df_Passenger_Miles_USA_clean['Total_regional']

df_Passenger_Miles_USA_clean['frac_trns_pkm_dem_regional_water_borne'] = df_Passenger_Miles_USA_clean['ferry_boat_data']/df_Passenger_Miles_USA_clean['Total_regional']


df_Passenger_Miles_USA_clean
```

```python
# The similarity of the obtained fractions with the information in https://www.bts.gov/sites/bts.dot.gov/files/states2020/Louisiana.pdf
# in the 'HOW RESIDENTS GET TO WORK' graph allow us to believe that the fractions of passenger per Km (private and pblic) is similar to the National numbers. 

years_proj = range(max(years_hist)+1,2050+1)


for var in frac_trns_pkm_dem_private_and_public_vars:
    
    df_historical = df_Passenger_Miles_USA_clean[['Year',var]].copy()
    df_historical['iso_code3'] = 'LA'
    df_historical['Region'] = 'louisiana'
    df_historical.to_csv(grandparentdir+'/'+var+'/input_to_sisepuede/historical/'+var+'.csv')

    # Projected value will be tha same as the last historical value

    year_val = []
    
    for year in years_proj:

        val = df_historical.loc[df_historical['Year']==max(years_hist)][var].values[0]

        year_val.append(['LA','louisiana',year,val])
        
    df_projected_out = pd.DataFrame(year_val, columns=['iso_code3','Region', 'Year',var])

    df_projected_out.to_csv(grandparentdir+'/'+var+'/input_to_sisepuede/projected/'+var+'.csv')
    
    print('Ready: ', var)
```

```python
# Regional fractions will be taken a as the national fractions

years_proj = range(max(years_hist)+1,2050+1)


for var in frac_trns_pkm_dem_regional_vars:
    
    df_historical = df_Passenger_Miles_USA_clean[['Year',var]].copy()
    df_historical['iso_code3'] = 'LA'
    df_historical['Region'] = 'louisiana'
    df_historical.to_csv(grandparentdir+'/'+var+'/input_to_sisepuede/historical/'+var+'.csv')

    # Projected value will be tha same as the last historical value

    year_val = []
    
    for year in years_proj:

        val = df_historical.loc[df_historical['Year']==max(years_hist)][var].values[0]

        year_val.append(['LA','louisiana',year,val])
        
    df_projected_out = pd.DataFrame(year_val, columns=['iso_code3','Region', 'Year',var])

    df_projected_out.to_csv(grandparentdir+'/'+var+'/input_to_sisepuede/projected/'+var+'.csv')
    
    print('Ready: ', var)
```
